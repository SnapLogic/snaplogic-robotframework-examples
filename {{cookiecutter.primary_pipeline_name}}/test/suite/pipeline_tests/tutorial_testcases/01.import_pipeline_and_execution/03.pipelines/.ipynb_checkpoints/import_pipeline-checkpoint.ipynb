# Import Pipeline — Interactive Notebook

Companion notebook for [`import_pipeline.robot`](./import_pipeline.robot).
Covers **both** test cases from the Robot file:

1. **`Import Pipeline`** — uses `Import Pipelines From Template` to import with a `<name>_<unique_id>` suffix (parent + child rows)
2. **`Import existing child Pipeline Without Unique ID`** — uses `Import Pipeline With Original Name` to import with a fixed name (no suffix)

Both Robot keywords are thin wrappers around the same underlying API — this notebook unpacks that one HTTP call so you can see exactly what happens.

🔍 **Robot vs Notebook** — The Robot keywords also set suite variables like `${<pipeline_name>_snode_id}` for downstream task creation. This notebook stores those in a Python `pipeline_snode_ids` dict instead.

## What this notebook covers

1. **Setup** — load env vars, build session, fetch `ORG_SNODE_ID`
2. **Helpers** — generate unique id, load `.slp`, import, list, delete
3. **Test Case 1 / Row 1** — `prime_oracle_baseline_tests.slp` → `prime_oracle_baseline_tests3_<unique_id>`
4. **Test Case 1 / Row 2** — `oracle.slp` → `oracle_<unique_id>`
5. **Test Case 2** — `email.slp` → `email` (no suffix)
6. **Verify** — list imported pipelines
7. **Cleanup** — delete imported pipelines (optional)

## Prerequisites

- Start Jupyter via `make jupyter-start` (or `make start-services`).
- Required env vars (auto-loaded by the Setup cell from `/app/.env`):
  - `URL`, `ORG_ADMIN_USER`, `ORG_ADMIN_PASSWORD`, `ORG_NAME`
  - `PIPELINES_LOCATION_PATH` (e.g. `test10_project_space/sl_project`)
- `.slp` files exist under `/app/src/pipelines/`:
  - `prime_oracle_baseline_tests.slp`, `oracle.slp`, `email.slp`

## SnapLogic API used here

| Operation | Method | Endpoint |
|---|---|---|
| Get ORG_SNODE_ID | `GET`    | `/api/1/rest/asset/list/{ORG}` → `entries[0].parent_snode_id` |
| Import pipeline  | `POST`   | `/api/2/{ORG_SNODE_ID}/rest/project/import/pipe/slp/{ORG}/{path}?duplicate_check=false` |
| List pipelines   | `GET`    | `/api/1/rest/asset/list/{ORG}/{path}` (filter `asset_type="Pipeline"`) |
| Delete pipeline  | `DELETE` | `/api/1/rest/asset/{ORG}/{path}/{pipeline_name}` |

## Section 1 — Setup

Load credentials, build a `requests.Session` with Basic Auth, and fetch
`ORG_SNODE_ID` (the import endpoint embeds it in the URL path).

````python
import os
import json
import time
import secrets
from pathlib import Path
import requests
from dotenv import load_dotenv

# --- Load env vars from root .env --------------------------------------------
load_dotenv("/app/.env", override=False)

# --- SnapLogic API credentials -----------------------------------------------
SL_URL  = os.environ.get("URL", "https://elastic.snaplogic.com")
SL_USER = os.environ.get("ORG_ADMIN_USER")
SL_PASS = os.environ.get("ORG_ADMIN_PASSWORD")
ORG     = os.environ.get("ORG_NAME")
PIPELINES_LOCATION_PATH = os.environ.get("PIPELINES_LOCATION_PATH")

missing = [k for k, v in {
    "ORG_ADMIN_USER": SL_USER, "ORG_ADMIN_PASSWORD": SL_PASS,
    "ORG_NAME": ORG, "PIPELINES_LOCATION_PATH": PIPELINES_LOCATION_PATH,
}.items() if not v]
if missing:
    raise RuntimeError(f"Missing required env vars: {missing}. Source your .env files first.")

session = requests.Session()
session.auth = (SL_USER, SL_PASS)
session.headers.update({"Content-Type": "application/json"})

# --- Fetch ORG_SNODE_ID (embedded in the import URL path) --------------------
resp = session.get(f"{SL_URL}/api/1/rest/asset/list/{ORG}")
resp.raise_for_status()
ORG_SNODE_ID = resp.json()["response_map"]["entries"][0]["parent_snode_id"]

print(f"✅ SnapLogic session ready")
print(f"   URL:                     {SL_URL}")
print(f"   Org:                     {ORG}")
print(f"   User:                    {SL_USER}")
print(f"   PIPELINES_LOCATION_PATH: {PIPELINES_LOCATION_PATH}")
print(f"   ORG_SNODE_ID:            {ORG_SNODE_ID}")
````

## Section 2 — Helper functions

These mirror the Robot keywords:

| Notebook helper | Robot keyword |
|---|---|
| `get_unique_id()` | `Get Unique Id` |
| `load_slp(slp_filename)` | `Load Json From File` |
| `build_payload(slp, pipeline_name)` | inline payload construction |
| `import_pipeline(slp_filename, pipeline_name, path, duplicate_check)` | `Import Pipeline` |
| `import_from_template(unique_id, path, pipeline_name, slp_filename, dup)` | `Import Pipelines From Template` |
| `import_with_original_name(path, pipeline_name, slp_filename, dup)` | `Import Pipeline With Original Name` |
| `list_pipelines(path)` / `pipeline_exists(...)` / `delete_pipeline(path, name)` | (read-only / cleanup) |

We also keep a `pipeline_snode_ids` dict to mimic the suite variables the Robot keywords set.

````python
# Where checked-in .slp files live inside the tools/jupyter container
PIPELINE_PAYLOAD_PATH = Path("/app/src/pipelines")

# Mimics the Robot suite variables ${<pipeline_name>_snode_id}
pipeline_snode_ids: dict[str, str] = {}


def get_unique_id() -> str:
    """Generate a unique suffix for pipeline names — mirrors Robot's `Get Unique Id`.

    Timestamp + short random hex keeps it short and human-readable while still
    avoiding collisions across parallel runs.
    """
    return f"{int(time.time())}_{secrets.token_hex(3)}"


def load_slp(slp_filename: str) -> dict:
    """Load a `.slp` file (plain JSON) from `/app/src/pipelines/`."""
    return json.loads((PIPELINE_PAYLOAD_PATH / slp_filename).read_text())


def build_payload(slp: dict, pipeline_name: str) -> dict:
    """Wrap the .slp contents in the import envelope and set the pipeline name.

    Mirrors the inline payload construction inside the Robot `Import Pipeline` keyword:
    `{"pipe": <slp>, ...}` with `pipe.property_map.info.label.value = pipeline_name`.
    """
    payload = {"pipe": slp}
    payload["pipe"]["property_map"]["info"]["label"]["value"] = pipeline_name
    return payload


def import_pipeline(
    slp_filename: str,
    pipeline_name: str,
    project_path: str,
    duplicate_check: bool = False,
) -> str:
    """Core import — POST the .slp payload to the import endpoint.

    Returns the new pipeline's `snode_id` (the API returns it as `response_map`).
    Raises if the API returns non-2xx (including when `duplicate_check=True` and
    the pipeline already exists).
    """
    slp = load_slp(slp_filename)
    payload = build_payload(slp, pipeline_name)
    url = f"{SL_URL}/api/2/{ORG_SNODE_ID}/rest/project/import/pipe/slp/{ORG}/{project_path}"
    resp = session.post(url, params={"duplicate_check": str(duplicate_check).lower()}, json=payload)
    if not resp.ok:
        print(f"❌ {resp.status_code} response body: {resp.text[:600]}")
    resp.raise_for_status()
    snode_id = resp.json().get("response_map")
    return snode_id


def import_from_template(
    unique_id: str,
    project_path: str,
    pipeline_name: str,
    slp_filename: str,
    duplicate_check: bool = False,
) -> str:
    """Mirrors Robot's `Import Pipelines From Template` — appends `_<unique_id>` to the name."""
    full_name = f"{pipeline_name}_{unique_id}"
    print(f"📦 Importing {slp_filename} → {full_name}  @ {ORG}/{project_path}")
    snode_id = import_pipeline(slp_filename, full_name, project_path, duplicate_check)
    pipeline_snode_ids[full_name] = snode_id
    print(f"✅ Imported [{full_name}]  snode_id={snode_id}")
    return snode_id


def import_with_original_name(
    project_path: str,
    pipeline_name: str,
    slp_filename: str,
    duplicate_check: bool = False,
) -> str:
    """Mirrors Robot's `Import Pipeline With Original Name` — keeps the name exactly as-is."""
    print(f"📦 Importing {slp_filename} → {pipeline_name}  @ {ORG}/{project_path}")
    snode_id = import_pipeline(slp_filename, pipeline_name, project_path, duplicate_check)
    pipeline_snode_ids[pipeline_name] = snode_id
    print(f"✅ Imported [{pipeline_name}]  snode_id={snode_id}")
    return snode_id


def list_pipelines(project_path: str) -> list[dict]:
    """List Pipeline assets under <ORG>/<project_path>."""
    url = f"{SL_URL}/api/1/rest/asset/list/{ORG}/{project_path}"
    r = session.get(url)
    r.raise_for_status()
    entries = r.json().get("response_map", {}).get("entries", [])
    return [e for e in entries if e.get("asset_type") == "Pipeline"]


def pipeline_exists(project_path: str, pipeline_name: str) -> dict | None:
    """Return the matching pipeline entry if it exists, else None."""
    for entry in list_pipelines(project_path):
        if entry["name"] == pipeline_name:
            return entry
    return None


def delete_pipeline(project_path: str, pipeline_name: str) -> None:
    """DELETE a pipeline using its path-based URL (same shape as the account API)."""
    url = f"{SL_URL}/api/1/rest/asset/{ORG}/{project_path}/{pipeline_name}"
    r = session.delete(url)
    if not r.ok:
        print(f"❌ {r.status_code} response body: {r.text[:300]}")
    r.raise_for_status()


# Generate one unique_id for this notebook run — mirrors the Robot suite setup
unique_id = get_unique_id()
print(f"✅ Helper functions defined")
print(f"   unique_id for this run: {unique_id}")
````

## Section 3 — Test Case 1 / Row 1: `prime_oracle_baseline_tests.slp`

Mirrors this Robot test row from `Import Pipeline`:

```robot
${unique_id}    ${PIPELINES_LOCATION_PATH}    prime_oracle_baseline_tests3    prime_oracle_baseline_tests.slp
```

Pipeline is imported as `prime_oracle_baseline_tests3_<unique_id>`. Each notebook run produces a fresh name (no collisions).

````python
import_from_template(
    unique_id=unique_id,
    project_path=PIPELINES_LOCATION_PATH,
    pipeline_name="prime_oracle_baseline_tests3",
    slp_filename="prime_oracle_baseline_tests.slp",
)
````

## Section 4 — Test Case 1 / Row 2: `oracle.slp`

Mirrors this Robot test row:

```robot
${unique_id}    ${PIPELINES_LOCATION_PATH}    ${pipeline_slp_name}    ${pipeline_name_for_imported_slp_file}
```

Where `${pipeline_slp_name}` = `oracle` and `${pipeline_name_for_imported_slp_file}` = `oracle.slp`. Pipeline is imported as `oracle_<unique_id>`.

````python
import_from_template(
    unique_id=unique_id,
    project_path=PIPELINES_LOCATION_PATH,
    pipeline_name="oracle",
    slp_filename="oracle.slp",
)
````

## Section 5 — Test Case 2: Import with original name (no suffix)

Mirrors the second Robot test case `Import existing child Pipeline Without Unique ID`:

```robot
${PIPELINES_LOCATION_PATH}    email    email.slp
```

Pipeline is imported as `email` exactly — no suffix. With `duplicate_check=False`
(default), re-running this cell **overwrites** any existing `email` pipeline
(preserving its `snode_id`, so downstream tasks keep working).

The commented-out Robot row demonstrates `duplicate_check=true`, which makes
the API return an error if the pipeline already exists — useful as a skip-if-exists guard.

````python
import_with_original_name(
    project_path=PIPELINES_LOCATION_PATH,
    pipeline_name="email",
    slp_filename="email.slp",
)

# Mirrors the commented-out row:
#   ${PIPELINES_LOCATION_PATH}    email    email.slp    duplicate_check=true
# Uncomment to demonstrate the skip-if-exists guard (will raise HTTPError if email exists):
#
# import_with_original_name(
#     project_path=PIPELINES_LOCATION_PATH,
#     pipeline_name="email",
#     slp_filename="email.slp",
#     duplicate_check=True,
# )
````

## Section 6 — Verify (list imported pipelines)

Lists all pipelines under `PIPELINES_LOCATION_PATH` and highlights the ones
we just imported. Also dumps the `pipeline_snode_ids` dict — these are the
snode_ids you'd hand off to a triggered-task creation step.

````python
imported_names = set(pipeline_snode_ids.keys())
all_pipes = list_pipelines(PIPELINES_LOCATION_PATH)

print(f"📋 Pipelines in {ORG}/{PIPELINES_LOCATION_PATH}: ({len(all_pipes)} total)")
for entry in sorted(all_pipes, key=lambda e: e["name"]):
    asset_id = (entry.get("asset_id") or "")[:16]
    marker = " ⭐ imported above" if entry["name"] in imported_names else ""
    print(f"   - {entry['name']:55s}  ({asset_id}...){marker}")

print(f"\n📌 pipeline_snode_ids (mirrors Robot's ${{<name>_snode_id}} suite vars):")
for name, snode_id in pipeline_snode_ids.items():
    print(f"   {name:55s} → {snode_id}")

found = {e["name"] for e in all_pipes}
missing = imported_names - found
if missing:
    print(f"\n⚠️  Imported but not found in list: {missing}")
else:
    print(f"\n✅ All {len(imported_names)} imported pipelines present in SnapLogic")
````

## Section 7 — Cleanup (optional)

Delete the pipelines created above so the project space is left clean.
Skip this if you want to inspect the pipelines in the SnapLogic Manager UI,
or if you plan to create triggered tasks against them next.

````python
for name in list(pipeline_snode_ids.keys()):
    if pipeline_exists(PIPELINES_LOCATION_PATH, name):
        delete_pipeline(PIPELINES_LOCATION_PATH, name)
        print(f"🗑️  deleted {PIPELINES_LOCATION_PATH}/{name}")
        del pipeline_snode_ids[name]
    else:
        print(f"ℹ️  {PIPELINES_LOCATION_PATH}/{name} not found — nothing to delete")

print(f"\n✅ Cleanup done — pipeline_snode_ids now: {pipeline_snode_ids}")
````

## Caveats vs the Robot keywords

This notebook is a simplified educational version. The Robot keywords additionally:

- Set **suite-level variables** like `${<pipeline_name>_snode_id}` that downstream Robot keywords (`Create Triggered Task`) consume automatically
- Auto-detect auth mode (Basic vs JWT vs OAuth2) and refresh tokens
- Provide structured Robot-style `Log` output for failure diagnosis
- Integrate with the broader suite's `Initialize Variables` keyword for `${unique_id}` generation

For production tests, **use the Robot keywords**. This notebook is for learning what's underneath.
